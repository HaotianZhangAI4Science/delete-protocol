# Processing after download

```python
#@title Write a bash for processing
com_file = open('1.sh','w')
com_file.write('''
#!/bin/bash

# 进入每个子文件夹
for dir in */; do
  cd "$dir"
  # 进入内部文件夹
  for inner_dir in */; do
    cd "$inner_dir"
    # 解压所有.gz文件
    for file in *.gz; do
      gunzip "$file"
    done
    # 复制所有解压后的文件到上层文件夹
    mv *.sdf ../../
    cd ../
  done
  cd ../
done

# 合并所有文件
cat *.sdf > all_zinc.sdf   
''')
com_file.close()
```

```python
from glob import glob
import os.path as osp
import pandas as pd
dirs = glob('./library/Zinc-Million-main/*')
total_smis = []
total_zinc_id = []
for dir in dirs:
    smi_files = glob(osp.join(dir, '*.smi'))
    for smi_file in smi_files:
        smi_content = pd.read_csv(smi_file, sep=' ')
        total_smis.extend(smi_content['smiles'].to_list())
        total_zinc_id.extend(smi_content['zinc_id'].to_list())
total_content = pd.DataFrame({'smiles':total_smis, 'zinc_id':total_zinc_id})
total_content.to_csv('./zinc.smi', sep=' ', index=False)
total_content = pd.read_csv('./zinc.smi', sep=' ')
```
